```python
import nbformat
from nbconvert import PythonExporter
from pathlib import Path
from traitlets.config import Config
```

```python
notebook_path = Path('/home/users/mendrika/ASNIE/Analysis/Core-RoA-Objects.ipynb')
export_path = notebook_path.with_suffix('')
```

```python
# Load notebook
with open(notebook_path, "r") as f:
    nb = nbformat.read(f, as_version=4)

# Filter out markdown cells
nb.cells = [cell for cell in nb.cells if cell.cell_type != "markdown"]

# Configure exporter to exclude input/output prompts
cfg = Config()
cfg.PythonExporter.exclude_input_prompt = True
cfg.PythonExporter.exclude_output_prompt = True

exporter = PythonExporter(config=cfg)
script, _ = exporter.from_notebook_node(nb)

# Save to .py file
with open(export_path.with_suffix('.py'), "w") as f:
    f.write(script)
```
