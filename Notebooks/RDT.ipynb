```python
import re

import xarray as xr
import pandas as pd

from pathlib import Path
```

```python
BASE = Path("/gws/ssde/j25b/swift/earajr/NWCSAF_archive/netcdf_2018/RDT")

def get_rdt_file(year, month, day, hour, minute, domain="WestAfrica"):
    date = f"{year:04d}{month:02d}{day:02d}"
    time = f"{hour:02d}{minute:02d}00"

    filename = f"S_NWC_RDT-CW_MSG4_{domain}-VISIR_{date}T{time}Z.nc"

    return BASE / date / filename
```

```python
f = get_rdt_file(2021, 4, 24, 22, 15)
ds = xr.open_dataset(f)
```

```python
df = pd.DataFrame({
    "lat": ds["LatIncG"].values,
    "lon": ds["LonIncG"].values,
    "area": ds["Surface"][:, 0].values,
    "btmin": ds["BTmin"][:, 0].values,
    "rain": ds["CRainRate"][:, 0].values,
    "speed": ds["MvtSpeed"].values,
    "dir": ds["MvtDirection"].values,
    "conv": ds["ConvType"].values,
})
```
